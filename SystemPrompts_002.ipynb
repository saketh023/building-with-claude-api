```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    user_message = {'role': 'user', 'content': text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {'role': 'assistant', 'content': text}
    messages.append(assistant_message)

def chat(messages, system = None, temperature=1.0):
    params={
        "model" : model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
    }
    
    if system:
        params["system"] = system
    
    response = client.messages.create(**params)
    return response.content[0].text
```

```python
messages = []
system_prompt = """
You are a patient math tutor. Do not directly answer a student's questions. Guide them to a solution step by step.
"""

add_user_message(messages, "How do I solve 4x+3 = 8 for x?")
answer = chat(messages, system=system_prompt)

print(answer)
```

```python
coding_msgs = []

coder_system_prompt="""
You are a proficient python developer. Write simple and understandable code without any explanation or comments. Provide only the python code to a given question.
"""

add_user_message(coding_msgs, "Write a python function that checks a string for duplicate characters.")

coding_ans = chat(coding_msgs, coder_system_prompt)

print(coding_ans)
```

```python
# temperature 
temp_msgs = []
add_user_message(temp_msgs, "Generate a one line movie idea.")
temp_ans = chat(temp_msgs, temperature=0.9)
print(temp_ans)
```
